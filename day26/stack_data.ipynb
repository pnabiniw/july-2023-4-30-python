```python
filt = df["Age"] > 40
df.loc[filt, ["Age", "Country", "Student", "LanguageWorkedWith"]]
```

```python
filt = (df["Country"] == "India") | (df["Country"] == "Canada")
df.loc[filt, ["Country", "Student", "LanguageWorkedWith"]]
# df[filt]
```

```python
countries = ["India", "United Kingdom", "Canada", "Thailand"]
filt = df["Country"].isin(countries)
df.loc[filt, ["Country", "Student", "LanguageWorkedWith"]]
```

```python
filt = df["Country"].isin(countries) & df['LanguageWorkedWith'].str.contains("Python")
df.loc[filt, ["Country", "Student", "LanguageWorkedWith"]]
```
